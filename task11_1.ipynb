```python
# Построить график

from sympy.plotting import plot
from sympy import Symbol, solve, evalf, solveset, diff, S, pprint
from sympy import*

x = Symbol('x') 
f = sin(x)**2 - cos(x)**2
p1 = plot(f, show = False)
p1.show()
```

```python
# Определить корни
ans = solve(f,x)
print(ans)
for i in ans:
    print(i.evalf())
```

```python
# Определить промежутки, на котором f > 0
pprint(solveset(f > 0, x, domain=S.Reals), use_unicode=True)
```

```python
# Определить промежутки, на котором f < 0
pprint(solveset(f < 0, x, domain=S.Reals), use_unicode=True)
```

```python
# Вычислить вершину

diff_f = diff(f,x)
print(diff_f)
solve(diff_f,x)
```

```python
# Найти интервалы, на которых функция возрастает
pprint(solveset(diff_f > 0, x, domain=S.Reals), use_unicode=True)
```

```python
# Найти интервалы, на которых функция убывает
pprint(solveset(diff_f < 0, x, domain=S.Reals), use_unicode=True)
```
