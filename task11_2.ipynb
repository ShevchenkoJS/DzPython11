# Определим корни

```python
from sympy.plotting import plot
from sympy import Symbol, solve, evalf, solveset, diff, S, pprint
from sympy import*

x = Symbol("x")
y = -12*x**4
solve(y)
```

```python
x = Symbol("x")
y = sin(cos(x))
solve(y)
```
